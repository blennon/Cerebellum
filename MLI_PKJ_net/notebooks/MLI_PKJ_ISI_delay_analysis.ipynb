<h1>MLI->PKJ ISI delay analysis</h1>

The purpose of this notebook is to analyze the data output by the scripts in ./experiments.  Basically, we connected an MLI up to a PKJ and simulated the effect of MLI feedforward inhibition on the PKJ ISI.  Below we show that the maximum synaptic conductance of the MLI to PKJ synapse is proportional to the ISI.

```python
import cPickle
sys.path.append('../../')
from MLI_PKJ_net import *
from brian import *
```

```python
def load_data(in_dir):
    '''
    loads the data ouput by ./experiments/MLI_PKJ_ISI_delay.py and ./experiments/MLI_PKJ_ISI_delay_rand_nS.py
    '''
    traces = cPickle.load(open(in_dir+'traces.pkl'))
    isis = cPickle.load(open(in_dir+'isis.pkl'))
    spikes = cPickle.load(open(in_dir+'spikes.pkl'))
    return traces, isis, spikes
```

```python
def draw_smooth_spikes(trace, ap = array([ 0.025,  .2,  0.7, 1.,  0.65, .05])):
    '''
    replaces the value of 0 in trace arrays with the waveform 'ap'
    '''
    new_trace = []
    last_v = None
    for v in trace:
        if v == 0:
            new_trace += list(ap*(0-last_v) + last_v)
            continue
        new_trace.append(v)
        last_v = v
    return array(new_trace)
```

```python
def plot_aligned_traces(traces, spikes, n=30, delay=12, ax=None, **plotoptions):
    '''
    plots all of the membrane potential traces from an experiment and aligns the first spike
    '''
    if ax is None:
        ax = subplot(111)
    for i in range(n):
        trace = extract_two_spike_trace(traces[i], spikes[i], float(10*ms), float(.25*ms))
        trace = draw_smooth_spikes(trace)
        plot(.25*ms*arange(trace.shape[0])*1000,trace*1000,**plotoptions)
    simpleaxis(ax)
    if delay is not None:
        mli_spike_time = argmax(trace)*.25+delay
        ax.annotate('',xy=(mli_spike_time,-55),xytext=(mli_spike_time-delay/1.5,-45),arrowprops=dict(facecolor='k', shrink=0.05),fontsize=16, color='k')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    xlim([-5,60])
    ylim([-75,0])
    add_scalebar(ax)
```

```python
def cm2inch(cm):
    '''convert cm to inches'''
    return cm*0.393701
```

<h2> Figures </h2>

```python
ISI_delay_data_dir_control = './' # the directory containing the output of ./experiments/MLI_PKJ_ISI_delay.py for the control case,
                                  # where the MLI->PKJ synaptic conductance is set to 0 nS
ISI_delay_data_dir_4nS = './' # the directory containing the output of ./experiments/MLI_PKJ_ISI_delay.py for the case
                                  # where the MLI->PKJ synaptic conductance is set to 4 nS
ISI_delay_rand_nS_data_dir = './' # the directory containing the output of ./experiments/MLI_PKJ_ISI_delay_rand_nS.py
```

```python
# comparison to non FFI case
fig=figure(figsize=(12,10), dpi=600)
traces_control, isis_control, spikes_control = load_data(ISI_delay_data_dir_control)
traces_4nS, isis_4nS, spikes_4nS = load_data(ISI_delay_data_dir_4nS)

ax = fig.add_subplot(311)
plot_aligned_traces(traces_control, spikes_control, 30, None, ax, color='#0B486B')


ax = fig.add_subplot(312)
plot_aligned_traces(traces_4nS, spikes_4nS, 30, delay=12, ax=ax, color='#3299BB')

ax = fig.add_subplot(325)
ax.hist(array(isis_control)*1000,15,alpha=.4,color='#0B486B')
ax.hist(array(isis_4nS)*1000,15,alpha=.4, color='#3299BB')
simpleaxis(ax)
tick_params(labelsize=16)
xlabel('ISI (ms)',fontsize=16)
ylabel('Count',fontsize=16)

ax = fig.add_subplot(326)
isis = cPickle.load(open(ISI_delay_rand_nS_data_dir+'isis.pkl'))
g_inhs = cPickle.load(open(ISI_delay_rand_nS_data_dir+'g_inh.pkl'))
g_inh = array([float(g)/1e-9 for g in g_inhs])
isi = array([float(i)/1e-3 for i in isis])
scatter(g_inh[g_inh<4], isi[g_inh<4],color='k')
xlim([0,4])
xlabel('Peak IPSC (nS)',fontsize=16)
ylabel('ISI (ms)',fontsize=16)
simpleaxis(ax)
fig.subplots_adjust(wspace=.5)
tick_params(labelsize=16)

tight_layout()
fig.savefig('ffi_delay_color.tiff',dpi=600*4.57/12)
fig.savefig('ffi_delay_color.png',dpi=600*4.57/12)
```
